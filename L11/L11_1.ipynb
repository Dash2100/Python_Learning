```python
import matplotlib.pyplot as plt
x = ['chinese', 'english', 'math', 'history', 'discover']
y1 = [1,5,6,8,5]
y2 = [2,6,4,8,6]
plt.figure(figsize=(8,5))
plt.bar(x,y1,align='edge',width=0.35,color='b')
plt.bar(x,y2,align='edge',width=-0.35,color='r')
plt.show()
```

```python
import random
Chinese = [random.randint(60,100) for _ in range(50)]
English = [random.randint(60,100) for _ in range(50)]
print(f"chinese={Chinese}")
print(f"english={English}")
```

```python
result = [chinese[i]+english[i] for i in range(50)]
print(f"總分: {result}")
```

```python
import numpy as np
ndChinese = np.array(chinese)
ndEnglish = np.array(English)
ndResult = ndChinese + ndEnglish
```

```python
ndResult
```

```python
xRange = range(50)
plt.figure(figsize=(8,5))
plt.plot(xRange,ndResult)
plt.show()
```
